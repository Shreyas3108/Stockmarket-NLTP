```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
df=pd.read_csv('Combined_News_DJIA.csv')
df.head()
```

```python
train=df[df['Date']<'2014-01-01']
test=df[df['Date']>'2014-01-01']
```

```python
trainlines = []
lentrain = len(train.index)
for row in range(0,lentrain):
    trainlines.append(' '.join(str(x) for x in train.iloc[row,2:27]))
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect = CountVectorizer()
basictrain = vect.fit_transform(trainlines)
print(basictrain.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Logreg=LogisticRegression()
```

```python
Logreg.fit(basictrain,train['Label'])
```

```python
testheadlines = []
for row in range(0,len(test.index)):
    testheadlines.append(' '.join(str(x) for x in test.iloc[row,2:27]))
basictest = vect.transform(testheadlines)
predictions = Logreg.predict(basictest)
```

```python
predictions
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(test['Label'],predictions)
```

```python
### 48% prediction
```

```python
basicwords = vect.get_feature_names()
basiccoeffs = Logreg.coef_.tolist()[0]
coeffdf = pd.DataFrame({'Word' : basicwords, 
                        'Coefficient' : basiccoeffs})
coeffdf = coeffdf.sort_values(['Coefficient', 'Word'], ascending=[0, 1])
coeffdf.head(25)
```

```python
coeffdf.tail(20)
```

```python
advvect = CountVectorizer(ngram_range=(2,2))
```

```python
advtrain = advvect.fit_transform(trainlines)
```

```python
advmodel = LogisticRegression()
```

```python
advmodel.fit(advtrain,train['Label'])
```

```python
testlines = []
lentest = len(test.index)
for row in range(0,lentest):
    testlines.append(' '.join(str(x) for x in test.iloc[row,2:27]))
```

```python
advtest = advvect.transform(testlines)
advpred=advmodel.predict(advtest)
```

```python
metrics.accuracy_score(test['Label'],advpred)
```

```python
advwords=advvect.get_feature_names()
advcoeff=advmodel.coef_.tolist()[0]
advdfcoef=pd.DataFrame({'Words':advwords,'Coeffiecient':advcoeff})
```

```python
advdfcoef=advdfcoef.sort_values(['Coeffiecient','Words'],ascending=[0,1])
```
